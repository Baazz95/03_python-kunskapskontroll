```python
import numpy as np
import matplotlib.pyplot as plt

#imported np and matplotlib
```

```python
#Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

#The two ways to plot in matplotlib are through "object-orientated (OO) style", 
#and utilizing pyplot to generate figures, axes, and plotting functions. 
#The recommended approach is using the OO style, as it provides more control over
#the production of the graphics. 
```

```python
#Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

#A figure could best be described as the foundation on which a visualization is built upon.
#Axes refer to the plotting area of a visualization.
#An axis refers to a specific x and y (horizontal and vertical) lines that are defined on a visualization.  
#Artists refer to any 'objects' that can be modified and are placed on the 'figure'.
```

```python
#When plotting in Matplotlib, what is the expected input data type?

#Numpy arrays are commonly used for plotting in matplotlib, as well as Pandas dataframe and series. 
```

```python
#Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. 
#Your plot should have a title and axis-labels.
```

```python
x = np.linspace(-4, 4, 50) #setting up 50 points between -4 and +4
y = x**2

figure, axes = plt.subplots() #setting up figure and axes

axes.plot(x, y)
axes.set_title("Plot of function y = x^2")
axes.set_xlabel("x axis")
axes.set_ylabel("y axis")

plt.show()
```

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())
```

```python
#need to create a figure for both of the scatter and bar plots.
figure, (axes1, axes2) = plt.subplots(1, 2, figsize=(12, 6)) #common figuresize for 2 plots is 12, 6.

#Setup for scatter plot:
axes1.scatter(random_data_x, random_data_y)
axes1.set_title("Scatter Plot")
axes1.set_xlabel("Random data x")
axes1.set_ylabel("Random data y")

#Setup for box plot:
axes2.bar(names, values)
axes2.set_title("Bar Plot (Fruit Data)")
axes2.set_xlabel("Fruit")
axes2.set_ylabel("Quantity")

plt.show()
```
